```python
def getDataset():
    data = pd.read_csv('data/text.csv',sep=',',encoding='utf-8')
    return data 

dataset = getDataset()
dataset.info()
dataset.head(10)
```

Давайте очистимо наші дані від тегів та розмітки, використаємо регулярні вирази. Також приведемо до нижнього регістру.

```python
import re

columns = ["Title","Body"]

def clearText(column):
    for i in range(0,1122):    
        dataset.loc[i][column] = re.sub(r"<[^>]+>", "", dataset.loc[i][column])
        dataset.loc[i][column] = re.sub('[^A-Za-zА-Яа-яҐґЄєІіЇї\']+', " ", dataset.loc[i][column])
        dataset.loc[i][column] = dataset.loc[i][column].lower()
clearText("Title")
clearText("Body")
```
