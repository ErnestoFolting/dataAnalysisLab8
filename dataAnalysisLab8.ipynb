```python
import pandas as pd 
import numpy as np
```

```python
def getDataset():
    data = pd.read_csv('data/text.csv',sep=',',encoding='utf-8')
    return data 

dataset = getDataset()
dataset.info()
dataset.head(10)
```

Давайте очистимо наші дані від тегів та розмітки, використаємо регулярні вирази. Також приведемо до нижнього регістру.

```python
import re

columns = ["Title","Body"]

def clearText(column):
    for i in range(0,1122):    
        dataset.loc[i][column] = re.sub(r"<[^>]+>", "", dataset.loc[i][column])
        dataset.loc[i][column] = re.sub('[^A-Za-zА-Яа-яҐґЄєІіЇї\']+', " ", dataset.loc[i][column])
        dataset.loc[i][column] = dataset.loc[i][column].lower()
clearText("Title")
clearText("Body")
```

```python
dataset.info()
dataset.head(10)
```

Давайте тепер зчитаємо стоп-слова, щоб потім не брати їх до аналізу.Також додамо свої стоп-слова, а саме різні абревіатури. 

```python
import csv
def getStop():
    stop = list()
    with open('data/stopWords.csv',encoding = 'utf-8') as stopCsv:
        reader = csv.reader(stopCsv,delimiter = ',')
        for row in reader:
            stop.append(row[0])
    return stop 

stop = getStop()

abbreviations = ['тис', 'млн', 'грн', 'т.я', 'вул', 'cек', 'хв', 'обл', 'кв', 'пл', 'напр', 'гл', 'і.о', 'зам']

for i in range(0,14):
    stop.append(abbreviations[i])
```

Давайте розіб'ємо текст на токени.

```python
import nltk

tokens = nltk.tokenize.word_tokenize(dataset["Body"][0])
def tokenize(column):
    for i in range(0,1122):    
        tokens = nltk.word_tokenize(dataset.loc[i][column])
        tokens = [word for word in tokens if word not in stop]
        dataset[column].loc[i] = tokens
tokenize("Title")
tokenize("Body")


dataset.info()
dataset.head(10)
```

Тепер давайте проведемо лематизацію.

```python
import pymorphy2

morph = pymorphy2.MorphAnalyzer(lang='uk')
def lematization(column):
    for i in range(0,1122):    
        lems = []
        for word in dataset.loc[i][column]:
            lem = morph.parse(word)[0].normal_form
            lems.append(lem)
        dataset.loc[i][column] = lems
lematization("Title")
lematization("Body")


dataset.info()
dataset.head(10)
```

Давайте напишемо функцію підрахунку кількостей слів в тексті.

```python
def wordCount(words):
    counts = {}
    for word in words:
        if word not in counts.keys():
            counts[word] = 1
        else:
            counts[word] += 1
    return dict(sorted(counts.items(), reverse=True, key=lambda x: x[1]))
print(wordCount(dataset.loc[0]["Body"]))
```

Тепер давайте зберемо всі "Body" до купи та знайдемо 10 слів, що найчастіше зустрічаються в корпусі. Корпус розглядаємо як збірник всіх "Body".

```python
import collections 
allBodies = []
for i in range(0,1122):
    allBodies += dataset.iloc[i]["Body"]
topCountWords = dict(collections.Counter(allBodies).most_common(10))

print(topCountWords)

topCountWordsLst = list(topCountWords.keys())
print(topCountWordsLst)
```

Тепер давайте знайдемо значення TF для цих 10 слів.

```python
def findTF():
    tfLst = []
    for wordId in range(0,len(topCountWordsLst)):
        tf = []
        for i in range(0,1122):
            if(topCountWordsLst[wordId] in wordCount(dataset.loc[i]["Body"]).keys()):
                numberInBody = wordCount(dataset.loc[i]["Body"])[topCountWordsLst[wordId]]
                totalNumberInBody = len(dataset.loc[i]["Body"])
                tfValue = numberInBody/totalNumberInBody
            else:
                tfValue = 0
            tf.append(tfValue)
        tfLst.append(tf)
    return tfLst 
tfLst = findTF()
tfLst = pd.DataFrame(tfLst).T
tfLst.columns = topCountWordsLst

tfLst.info()
tfLst.head(15)
```

Знайдемо значення IDF для цих 10 слів.

```python
def findIDF():
    idf = []
    for wordId in range(0,len(topCountWordsLst)):
        count = 0
        for i in range(0,1122):
            if(topCountWordsLst[wordId] in wordCount(dataset.loc[i]["Body"]).keys()):
                count += 1
        idf.append(np.log(1122/count))
    return idf 
idf = findIDF()
idf = pd.DataFrame(idf).T
idf.columns = topCountWordsLst

idfLst = list(idf.loc[0])
print(idfLst)
```

Тепер знайдемо значення TF-IDF.

```python
def findTFIDF():
    tfIdf = []
    for i in range(0,1122):
        currTFIDF = []
        for j in range(0,10):
            currTFIDF.append(tfLst.loc[i][j]*idfLst[j])
        tfIdf.append(currTFIDF)
    return tfIdf

tfIdf = findTFIDF()
tfIdf = pd.DataFrame(tfIdf)
tfIdf.columns = topCountWordsLst
tfIdf.info()
tfIdf.head(10)
```

Тепер давайте створимо та наповнимо Bag of Word.

```python
allUniqueWords = list(wordCount(allBodies).keys())
allCounts = []
for j in range(0,1122):
    allCounts.append(wordCount(dataset.loc[j]["Body"]))

def createBagOfWord():
    arrays = []
    for i in range(0,len(allUniqueWords)):
        arr = []
        word = allUniqueWords[i]
        for j in range(0,1122):
            if word in allCounts[j].keys():
                arr.append(allCounts[j][word])
            else:
                arr.append(0)
        arrays.append(arr)
    return arrays

bag = createBagOfWord()
bag = pd.DataFrame(bag).T
bag.columns = allUniqueWords
bag.info()
bag.head(20)
```
